# Distribuzioni di probabilità

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

## Dist di Bernoulli

```python
dist = st.bernoulli(0.3)
```

La **distribuzione di Bernoulli** calcola le probabilità di un evento a due esiti con probabilità $p$ e $1-p$.

$X \sim B(p)$,  $supporto = D_X = \lbrace0, 1\rbrace$

Valore Attesto: $E(X) = p$

Varianza: $Var(X) = p(1-p)$

```python
# Probability Mass Function
x = np.arange(-1, 3, 1)
plt.vlines(x, 0, dist.pmf(x), linestyles='dotted')
plt.plot(x, dist.pmf(x), 'o')
plt.show()
```

```python
#Cumulative Distribution Function
plt.step(x, dist.cdf(x))
plt.show()
```

```python
# Definisci la distribuzione Bernoulli con parametro p=0.3
dist = st.bernoulli(0.3)

# Valori per l'asse x
x = np.arange(-1, 3, 1)

# Crea una figura con due sottotrame (subplot) affiancate
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Plot della PMF nella prima sottotrama
axes[0].vlines(x, 0, dist.pmf(x), linestyles='dotted')
axes[0].plot(x, dist.pmf(x), 'o')
axes[0].set_title('PMF')
axes[0].set_xlabel('x')
axes[0].set_ylabel('Probability')

# Plot della CDF nella seconda sottotrama
axes[1].step(x, dist.cdf(x),where='post')
axes[1].set_title('CDF')
axes[1].set_xlabel('x')
axes[1].set_ylabel('Cumulative Probability')
axes[0].set_xticks(x)
axes[1].set_xticks(x)
# Mostra il grafico
plt.tight_layout()
plt.show()
```

## Dist Binomiali

```python
dist = st.binom(10, 0.4)
```

La **distribuzione binomiale** rappresenta una serie di $n$ variabili aleatorie bernoulliane con stessa probabilità $p$ (o un evento bernoulliano ripetuto $n$ volte) di cui vogliamo sapere il numero di successi.

$X \sim B(n, p)$,  $supporto = D_X = \lbrace0, 1, 2, ..., \infty\rbrace$

Valore Atteso: $E(X) = np$

Varianza: $Var(X) = np(1-p)$

```python
# Probability Mass Function
x = np.arange(0, 11, 1)
plt.vlines(x, 0, dist.pmf(x), linestyles='dotted')
plt.plot(x, dist.pmf(x), 'o')
plt.show()
```

```python
#Cumulative Distribution Function
plt.step(x, dist.cdf(x))
plt.show()
```

La distribuzione binomiale inoltre gode della proprietà di **riproducibilità**.

**N.B.:** è sempre possibile vedere una sequenza di bernoulliane come una binomiale a patto che abbiano la stessa probabilità $p$.

```python
# Definisci la distribuzione binomiale con parametri n=10 e p=0.4
dist = st.binom(10, 0.4)

# Valori per l'asse x
x = np.arange(0, 11, 1)

# Crea una figura con due sottotrame (subplot) affiancate
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Plot della PMF nella prima sottotrama
axes[0].vlines(x, 0, dist.pmf(x), linestyles='dotted')
axes[0].plot(x, dist.pmf(x), 'o')
axes[0].set_title('PMF')
axes[0].set_xlabel('x')
axes[0].set_ylabel('Probability')

# Plot della CDF nella seconda sottotrama
axes[1].step(x, dist.cdf(x),where='post')
axes[1].set_title('CDF')
axes[1].set_xlabel('x')
axes[1].set_ylabel('Cumulative Probability')
axes[0].set_xticks(x)
axes[1].set_xticks(x)
# Mostra il grafico
plt.tight_layout()
plt.show()
```

## Dist di Poisson

```python
dist = st.poisson(3)
```

La **distribuzione di Poisson** esprime le probabilità che si verifichino un numero $n$ di elementi in un determinato lasso di tempo $t$ (**discreto!**) sapendo che mediamente se ne verificano $\lambda$.

$X \sim P(\lambda)$,  $supporto = D_X = \lbrace 0 \rbrace \cup \mathbb{N^+} $

Valore Atteso: $ E(X) = \lambda $

Varianza: $ Var(X) = \lambda $

```python
# Probability Mass Function
x = np.arange(0, 11, 1)
plt.vlines(x, 0, dist.pmf(x), linestyles='dotted')
plt.plot(x, dist.pmf(x), 'o')
plt.show()
```

```python
#Cumulative Distribution Function
plt.step(x, dist.cdf(x))
plt.show()
```

La distribuzione di Poisson inoltre gode della proprietà di **riproducibilità**.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

# Definisci la distribuzione di Poisson con parametro lambda=3
dist = st.poisson(3)

# Valori per l'asse x
x = np.arange(0, 11, 1)

# Crea una figura con due sottotrame (subplot) affiancate
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Plot della PMF nella prima sottotrama
axes[0].vlines(x, 0, dist.pmf(x), linestyles='dotted')
axes[0].plot(x, dist.pmf(x), 'o')
axes[0].set_title('PMF')
axes[0].set_xlabel('x')
axes[0].set_ylabel('Probability')

# Plot della CDF nella seconda sottotrama
axes[1].step(x, dist.cdf(x),where='post')
axes[1].set_title('CDF')
axes[1].set_xlabel('x')
axes[1].set_ylabel('Cumulative Probability')
axes[0].set_xticks(x)
axes[1].set_xticks(x)

# Mostra il grafico
plt.tight_layout()
plt.show()
```

## Dist Geometrica

```python
dist = st.geom(0.34)
```

La **distribuzione geometrica** serve per calcolare il numero di insuccessi prima di ottenere un successo(senza contarlo) ripetendo più volte un esperimento bernoulliano con probabilità $p$.

$X\sim G(p)$, $supporto = D_X = \lbrace 1, 2, ... \rbrace$

Valore Atteso: $E(X) = \frac{1-p}{p}$

Varianza: $ Var(X) = \frac{1-p}{p^2} $

```python
# Probability Mass Function
x = np.arange(1, 11, 1)
plt.vlines(x, 0, dist.pmf(x), linestyles='dotted')
plt.plot(x, dist.pmf(x), 'o')
plt.xticks(np.arange(1, 11, 1))
plt.yticks(np.arange(0, 0.8, 0.1))
plt.show()
```

```python
#Cumulative Distribution Function
plt.step(x, dist.cdf(x))
plt.show()
```

La distribuzione geometrica inoltre gode della proprietà di **assenza di memoria**.

```python

# Definisci la distribuzione geometrica con parametro p=0.34
dist = st.geom(0.34)

# Valori per l'asse x
x = np.arange(1, 11, 1)

# Crea una figura con due sottotrame (subplot) affiancate
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Plot della PMF nella prima sottotrama
axes[0].vlines(x, 0, dist.pmf(x), linestyles='dotted')
axes[0].plot(x, dist.pmf(x), 'o')
axes[0].set_title('PMF')
axes[0].set_xlabel('x')
axes[0].set_ylabel('Probability')
axes[0].set_xticks(np.arange(1, 11, 1))
axes[0].set_yticks(np.arange(0, 0.8, 0.1))

# Plot della CDF nella seconda sottotrama
axes[1].step(x, dist.cdf(x),where='post')
axes[1].set_title('CDF')
axes[1].set_xlabel('x')
axes[1].set_ylabel('Cumulative Probability')
axes[1].set_xticks(np.arange(1, 11, 1))

# Mostra il grafico
plt.tight_layout()
plt.show()
```

## Dist Uniforme Continua

```python
dist = st.uniform(1, 10) #bisogna dargli in pasto il valore +basso a e il valore +alto b
```

La **distribuzione uniforme continua** rappresenta un evento che è equiprobabile per tutti i punti appartenenti all'intervallo $[a, b]$.

$X \sim U(a, b)$, $supporto = D_X = [a, b]$

Valore Atteso: $E(X) = \frac{a+b}{2}$

Varianza: $ Var(X) = \frac{(b-a)^2}{12} $

```python
# Probability Density Function
x = np.arange(1, 11, 1)
plt.step(x, dist.pdf(x))
plt.xticks(np.arange(1, 12, 1))
plt.yticks(np.arange(0, 0.8, 0.1))
plt.show()
# devo farla così "ristretta" per farla uscire bene
```

```python
#Cumulative Distribution Function
plt.plot(x, dist.cdf(x))
plt.show()
```

```python
# Definisci la distribuzione uniforme continua nell'intervallo [1, 10]
dist = st.uniform(1, 10)

# Valori per l'asse x
x = np.arange(1, 11, 1)

# Crea una figura con due sottotrame (subplot) affiancate
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Plot della PDF nella prima sottotrama
axes[0].step(x, dist.pdf(x))
axes[0].set_title('PDF')
axes[0].set_xlabel('x')
axes[0].set_ylabel('Probability Density')
axes[0].set_xticks(np.arange(1, 12, 1))
axes[0].set_yticks(np.arange(0, 0.8, 0.1))

# Plot della CDF nella seconda sottotrama
axes[1].plot(x, dist.cdf(x))
axes[1].set_title('CDF')
axes[1].set_xlabel('x')
axes[1].set_ylabel('Cumulative Probability')
axes[1].set_xticks(np.arange(1, 12, 1))

# Mostra il grafico
plt.tight_layout()
plt.show()
```

## Dist Normale

```python
dist = st.norm(13, 44)
```

La **distribuzione normale** approssima variabili aleatore che tendono a concentrarsi attorno ad un valore medio $\mu$ con varianza $\sigma^2$.

$X \sim N(\mu, \sigma^2)$, $supporto = D_X = \mathbb{R}$

Valore Atteso: $ E(X) = \mu$

Varianza: $ Var(X) = \sigma^2 $

```python
# Probability Density Function
x = np.arange(-100, 126, 1)
plt.plot(x, dist.pdf(x))
plt.show()
```

```python
#Cumulative Distribution Function
plt.plot(x, dist.cdf(x))
plt.show()
```

**Normale standard**: distribuzione normale con $\mu = 0$ e $ \sigma = 1 $ 

```python
# Definisci la distribuzione normale con media=13 e deviazione standard=44
dist = st.norm(13, 44)

# Valori per l'asse x
x = np.arange(-100, 126, 1)

# Crea una figura con due sottotrame (subplot) affiancate
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Plot della PDF nella prima sottotrama
axes[0].plot(x, dist.pdf(x))
axes[0].set_title('PDF')
axes[0].set_xlabel('x')
axes[0].set_ylabel('Probability Density')

# Plot della CDF nella seconda sottotrama
axes[1].plot(x, dist.cdf(x))
axes[1].set_title('CDF')
axes[1].set_xlabel('x')
axes[1].set_ylabel('Cumulative Probability')

# Mostra il grafico
plt.tight_layout()
plt.show()
```
