```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import math
import statsmodels.api as sm
```

```python
def pdf(x, a):
    if x > 1:
        return 0
    elif x < 0:
        return 0
    else:
        return (a * x**2 + 1 - a / 3)

a = 5/2
x = np.arange(0, 1.1, 0.1)
y = [pdf(i, a) for i in x]

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('pdf(x;a)')
plt.title('Plot della Funzione di Densità di Probabilità')
plt.show()
```
