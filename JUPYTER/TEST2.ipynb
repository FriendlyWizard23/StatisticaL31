```python
import numpy as np
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
import math
```

```python
x=np.arange(0,100,0.05)
expon=st.expon(scale=1/(0.1))
plt.plot(x,expon.cdf(x))
plt.show()
```

```python
geom=st.geom(0.1)
plt.step(x,geom.cdf(x),where='post')
plt.show()
```
